```python
import tensorflow as tf
```

```python
#创建张量
a = tf.range(10*28*28*3)
new_a = tf.reshape(a,[10,28,28,3])
```

```python
#利用索引取出第二张图片 索引为1
with tf.Session() as sess:
    res_1 = sess.run(new_a[1])
    print(res_1)
```

```python
#利用切片取出第二张图片
b = tf.slice(new_a,[1,0,0,0],[1,28,28,3])
with tf.Session() as sess:
    res_2 = sess.run(b)
    print(res_2)
```

```python
#索引没有对原始的张量进行修改，而切片进行了修改
#索引0 2 4 6 即可取出第1 3 5 7张图片
```

```python
#切片取出第6-8张（包括6不包括8）图片中中心区域（14*14）的部分
c = tf.slice(new_a,[5,14,14,0],[2,1,1,3])
with tf.Session() as sess:
    res_3 = sess.run(c)
    print(res_3)
```

```python
#将图片切片，每一个通道为一张图片
new_a_1 = tf.slice(new_a,[0,0,0,0],[10,28,28,1])
new_a_2 = tf.slice(new_a,[0,0,0,1],[10,28,28,1])
new_a_3 = tf.slice(new_a,[0,0,0,2],[10,28,28,1])
with tf.Session() as sess:
    res_4 = sess.run(new_a_1)
    res_5 = sess.run(new_a_2)
    res_6 = sess.run(new_a_3)
    print(res_4)
    print(res_5)
    print(res_6)
```

```python
#`tf.shape(img)`返回的张量的阶数以及`shape`属性的值
rank = tf.shape(new_a)
print(tf.rank(rank))
print(tf.shape(new_a))
```
